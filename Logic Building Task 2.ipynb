# **Problem 1: Unique Words in a Sentence**

```python
# 1. Get the sentence input from the user
sentence = input("Enter sentence: ")

# 2. Split the sentence into individual words (Hint 1)
words_list = sentence.split()

# 3. Convert the list of words into a set to remove duplicates (Hint 2)
unique_words = set(words_list)

# 4. Print both the set and its length (Hint 3)
print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

# **Problem 2: Highest Salary from Employee Data**

```python
# 1. Store employee names and their salaries in a dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# 2. Determine which employee has the highest salary
highest_paid_employee = max(employees, key=employees.get)

# 3. Display the result
print(f"The employee with the highest salary is: {highest_paid_employee}")
```

```python
#Alternate Code

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_salary = 0
top_employee = ""

for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        top_employee = name

print(f"The employee with the highest salary is: {highest_paid_employee}")
print(f"Salary: {highest_salary}")
```

# **Problem 3: Find Maximum and Minimum Values**

```python
# 1. Define the input list
numbers = [45, 22, 89, 10, 66]

# 2. Initialize both maximum and minimum with the first element
maximum = numbers[0]
minimum = numbers[0]

# 3. Hint 2: Use a loop to compare each element
for num in numbers:
    # Update maximum if a larger number is found
    if num > maximum:
        maximum = num

    # Update minimum if a smaller number is found
    if num < minimum:
        minimum = num

# 4. Print results in the expected format
print(f"List: {numbers}")
print(f"Max: {maximum}")
print(f"Min: {minimum}")
```

# **Problem 4: Count Products Above a Price Threshold**

```python
# 1. Define the input list of prices
prices = [450, 1200, 899, 1500, 300]

# 2. Initialize a counter variable to keep track of products above 1000
count = 0

# 3. Use a loop to compare each price with 1000 (as per the hint)
for price in prices:
    if price > 1000:
        count += 1

# 4. Display the result in the expected format
print(f"Products above 1000: {count}")
```

# **Problem 5: Calculate Attendance Percentage**

```python
# 1. Define the input list
attendance = ["P", "P", "A", "P", "P"]

# 2. Hint 1: Count the number of "P" values
present_count = attendance.count("P")

# 3. Hint 2: Divide by the total number of days (the length of the list)
total_days = len(attendance)

# 4. Calculate the percentage
percentage = (present_count / total_days) * 100

# 5. Display the result in the expected format
print(f"Attendance Percentage: {percentage}")
```

# **Problem 7: Count Character Frequency**

```python
# 1. Define the input string
input_string = "pythonp"

# 2. Hint 1: Initialize an empty dictionary to store counts
char_counts = {}

# 3. Hint 2: Loop through each character and increment its count
for char in input_string:
    if char in char_counts:
        # If character is already in dictionary, add 1 to its current value
        char_counts[char] += 1
    else:
        # If character is not in dictionary, add it with a starting count of 1
        char_counts[char] = 1

# 4. Display the resulting dictionary
print(char_counts)
```

# **Problem 10: Calculate Average Marks**

```python
# 1. Define a list of marks
marks = [70, 75, 80, 65]

# 2. Hint 1: Calculate the sum of all values
total_sum = sum(marks)

# 3. Hint 2: Divide by the total number of marks
total_count = len(marks)
average = total_sum / total_count

# 4. Display the result in the expected format
print(f"Average Marks: {average}")
```
