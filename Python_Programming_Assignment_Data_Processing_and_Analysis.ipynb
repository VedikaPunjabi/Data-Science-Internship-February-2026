# **Problem Statement 1: Employee Performance Bonus Eligibility**

```python
# 1. Ask the user for the number of employees
n = int(input("Enter the number of employees: "))
employees = {}

for i in range(n):
    name = input(f"Enter name for employee {i+1}: ")
    score = int(input(f"Enter score for {name}: "))
    employees[name] = score

# --- Logic starts here ---
max_score = max(employees.values())
top_performers = [name for name, score in employees.items() if score == max_score]

print(f"\nTop Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {max_score})")
```

# **Problem Statement 2: Search Query Keyword Analysis**

```python
import string

# 1. Take search query from user
query = input("Enter your search query: ")

# --- Logic starts here ---
query_lower = query.lower()
translator = str.maketrans('', '', string.punctuation)
clean_query = query_lower.translate(translator)

words = clean_query.split()
word_counts = {}

for word in words:
    word_counts[word] = word_counts.get(word, 0) + 1

result = {word: count for word, count in word_counts.items() if count > 1}

print("\nKeywords searched more than once:")
print(result)
```

# **Problem Statement 3: Sensor Data Validation**

```python
# 1. Take list of numbers separated by space
user_input = input("Enter sensor readings separated by spaces: ")

# Convert the string input into a list of integers
sensor_readings = [int(x) for x in user_input.split()]

# --- Logic starts here ---
valid_readings = [(hour, value) for hour, value in enumerate(sensor_readings) if value % 2 == 0]

print("\nValid Sensor Readings (Hour, Value):")
print(valid_readings)
```

# **Problem Statement 4: Email Domain Usage Analysis**

```python
# Ask user for emails separated by commas or spaces
user_input = input("Enter email addresses separated by commas: ")
# Create list by splitting and removing extra whitespace
emails = [e.strip() for e in user_input.replace(',', ' ').split()]

domain_counts = {}
for email in emails:
    if '@' in email:
        domain = email.split('@')[1]
        domain_counts[domain] = domain_counts.get(domain, 0) + 1

total_users = len(emails)
print("\nDomain Usage Analysis:")
for domain, count in domain_counts.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage:.0f}%")
```

# **Problem Statement 5: Sales Spike Detection**

```python
# 1. Take list of sales numbers separated by space from the user
user_input = input("Enter daily sales figures separated by spaces: ")

# 2. Convert the string input into a list of integers
sales = [int(x) for x in user_input.split()]

# 3. Calculate the average daily sales
# We check if the list is not empty to avoid division by zero
if len(sales) > 0:
    average_sales = sum(sales) / len(sales)

    # 4. Define the spike threshold (30% above average)
    spike_threshold = average_sales * 1.30

    print(f"\nAnalysis Results:")
    print(f"Average Sales: {average_sales:.2f}")
    print(f"Spike Threshold (Average + 30%): {spike_threshold:.2f}")
    print("-" * 30)

    # 5. Detect and display days where sales are a spike
    found_spike = False
    for day, value in enumerate(sales, 1):
        if value > spike_threshold:
            print(f"Day {day}: {value}")
            found_spike = True

    if not found_spike:
        print("No sales spikes detected.")
else:
    print("No data entered.")
```

# **Problem Statement 6: Duplicate User ID Detection**

```python
# 1. Take user input
user_input = input("Enter user IDs separated by spaces or commas: ")

# 2. Clean and convert input into a list
user_ids = [uid.strip() for uid in user_input.replace(',', ' ').split() if uid.strip()]

# 3. Count frequencies
id_counts = {}
for uid in user_ids:
    id_counts[uid] = id_counts.get(uid, 0) + 1

# 4. Display results
print("\nDuplicate Detection Results:")
has_duplicates = False
for uid, count in id_counts.items():
    if count > 1:
        print(f"{uid} → {count} times")
        has_duplicates = True

if not has_duplicates:
    print("No duplicates found.")
```
