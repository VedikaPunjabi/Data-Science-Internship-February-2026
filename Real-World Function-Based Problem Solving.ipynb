# **Problem 1: Social Media – Post Engagement Analyzer**

```python
def analyze_engagement(likes_list):
    """
    Analyzes social media post engagement by summing up likes from a list.
    """
    total_likes = 0

    # Hint: Use loop for calculations
    for likes in likes_list:
        total_likes += likes

    # Rule: Determine status based on total likes
    if total_likes >= 1000:
        post_status = "Viral Post"
    else:
        post_status = "Normal Engagement"

    # Expected Output Format
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {post_status}")

# --- Test Cases ---

# Case 1: Post that becomes Viral
print("--- Case 1 ---")
engagement_data_1 = [400, 350, 300] # Total = 1050
analyze_engagement(engagement_data_1)

print("\n--- Case 2 ---")
# Case 2: Post with Normal Engagement
engagement_data_2 = [100, 200, 150] # Total = 450
analyze_engagement(engagement_data_2)
```

# **Problem 3: Agriculture – Rainfall Adequacy Checker**

```python
def check_rainfall_adequacy(rainfall_data, required_level):
    """
    Analyzes seasonal rainfall data to check if it meets a required threshold.
    """
    total_rainfall = 0

    # Hint: Use loop for calculations
    for rain in rainfall_data:
        total_rainfall += rain

    # Math operation: Calculate average rainfall
    if len(rainfall_data) > 0:
        average_rainfall = total_rainfall / len(rainfall_data)
    else:
        average_rainfall = 0

    # Rule: Determine status based on average vs required level
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    # Expected Output Format
    # Using :.0f to round to a whole number as shown in the example
    print(f"Average Rainfall: {average_rainfall:.0f}")
    print(f"Rainfall Status: {status}")

# --- Test Cases ---

# Case 1: Adequate Rainfall (Matching the example output)
print("--- Case 1 ---")
# Example data that averages to 72: (80+70+66+72)/4 = 72
rainfall_list_1 = [80, 70, 66, 72]
required_threshold = 70
check_rainfall_adequacy(rainfall_list_1, required_threshold)

print("\n--- Case 2 ---")
# Case 2: Inadequate Rainfall
rainfall_list_2 = [40, 50, 45, 35] # Average: 42.5
check_rainfall_adequacy(rainfall_list_2, 50)
```

# **Problem 4: Social Media – Duplicate Account Detection**

```python
def detect_duplicate_accounts(usernames):
    """
    Detects if there are duplicate usernames in a list using a set.
    """
    # Hint: Use set data structure
    # A set only contains unique elements
    unique_usernames = set(usernames)

    # Rule: If length of the set is less than the original list,
    # it means duplicates were removed.
    if len(usernames) != len(unique_usernames):
        duplicates_found = "Yes"
    else:
        duplicates_found = "No"

    # Expected Output Format
    print(f"Duplicate Accounts Found: {duplicates_found}")

# --- Test Cases ---

# Case 1: Duplicates exist (Matching the example output)
print("--- Case 1 ---")
user_list_1 = ["alice_99", "bob_dev", "alice_99", "charlie_k"]
detect_duplicate_accounts(user_list_1)

print("\n--- Case 2 ---")
# Case 2: All accounts are unique
user_list_2 = ["user1", "user2", "user3", "user4"]
detect_duplicate_accounts(user_list_2)
```

# **Problem 6: Agriculture – Premium Crop Price Filter**

```python
def filter_premium_crops(prices):
    """
    Filters a list of crop prices to identify those meeting the premium threshold.
    """
    # Hint: Use list filtering logic
    # We create a new list containing only prices > 2000
    premium_crops = []

    for price in prices:
        # Rule: Price > 2000 -> Premium Crop
        if price > 2000:
            premium_crops.append(price)

    # Expected Output Format
    print(f"Premium Crops: {premium_crops}")

# --- Test Case ---

# Example input containing a mix of regular and premium prices
crop_prices = [1500, 2500, 1800, 3200, 2000, 4100]

# Running the function
filter_premium_crops(crop_prices)
```

# **Problem 8: Banking – Daily Transaction Limit Checker**

```python
def validate_transaction(amount):
    """
    Validates a transaction amount against the daily limit of 50,000.
    """
    daily_limit = 50000

    # Hint: Use conditional logic to check the limit
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    # Expected Output Format
    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")

# --- Test Cases ---

# Case 1: Transaction exceeds limit (Matching the example output)
print("--- Case 1 ---")
validate_transaction(60000)

print("\n--- Case 2 ---")
# Case 2: Transaction within limit
validate_transaction(25000)

print("\n--- Case 3 ---")
# Case 3: Transaction at exactly the limit
validate_transaction(50000)
```

# **Problem 9: E-Learning – Student Attendance Eligibility System**

```python
def check_exam_eligibility(attendance_list):
    """
    Calculates attendance percentage from a list and determines exam eligibility.
    """
    days_present = 0
    total_days = len(attendance_list)

    # Hint: Use list and loop to count days present
    for status in attendance_list:
        if status == 1:
            days_present += 1

    # Calculate attendance percentage
    attendance_percentage = (days_present / total_days) * 100

    # Rule: Check eligibility based on the 75% threshold
    if attendance_percentage >= 75:
        eligibility_status = "Eligible"
    else:
        eligibility_status = "Not Eligible"

    # Expected Output Format
    print(f"Attendance Percentage: {attendance_percentage:.1f}")
    print(f"Exam Eligibility: {eligibility_status}")

# --- Test Cases ---

# Case 1: 80% Attendance (4 out of 5 days present)
print("--- Case 1 ---")
attendance_data_1 = [1, 1, 0, 1, 1]
check_exam_eligibility(attendance_data_1)

print("\n--- Case 2 ---")
# Case 2: 60% Attendance (3 out of 5 days present)
attendance_data_2 = [1, 0, 1, 0, 1]
check_exam_eligibility(attendance_data_2)
```

# **Problem 10: Smart Electricity Bill Analyzer**

```python
def analyze_electricity_bill(units):
    """
    Calculates the electricity bill based on slabs and classifies usage.
    """
    total_bill = 0

    # Hint: Use conditional logic for slabs
    if units <= 100:
        total_bill = units * 3
    elif units <= 200:
        # First 100 at ₹3, remaining units at ₹5
        total_bill = (100 * 3) + ((units - 100) * 5)
    else:
        # First 100 at ₹3, next 100 at ₹5, remaining units at ₹7
        total_bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage Classification Logic
    if total_bill < 500:
        usage_status = "Low Usage"
    elif 500 <= total_bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    # Hint: Use return values
    return total_bill, usage_status

# --- Test Cases ---

# Test Case 1: Low Usage
units_consumed = 80
bill, status = analyze_electricity_bill(units_consumed)
print(f"Units: {units_consumed} | Total Bill: ₹{bill} | Status: {status}")

# Test Case 2: Moderate Usage
units_consumed = 250 # (100*3) + (100*5) + (50*7) = 300 + 500 + 350 = 1150
bill, status = analyze_electricity_bill(units_consumed)
print(f"Units: {units_consumed} | Total Bill: ₹{bill} | Status: {status}")

# Test Case 3: High Usage
units_consumed = 400 # (100*3) + (100*5) + (200*7) = 300 + 500 + 1400 = 2200
bill, status = analyze_electricity_bill(units_consumed)
print(f"Units: {units_consumed} | Total Bill: ₹{bill} | Status: {status}")
```
