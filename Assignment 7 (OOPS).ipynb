# **Problem 1: Smart Light Controller**

```python
class SmartLight:
    def __init__(self, name):
        """Initialize the light with a name and a default status of OFF."""
        self.name = name
        self.status = "OFF"

    def set_action(self, action):
        """Update the status of the light based on the given action (ON/OFF)."""
        action = action.upper()
        if action == "ON" or action == "OFF":
            self.status = action
        else:
            print("Invalid action. Please use 'ON' or 'OFF'.")

    def display_status(self):
        """Print the current status of the light in the required format."""
        print(f"{self.name} is {self.status}")

# --- Testing with the Sample Input ---

# 1. Getting Input (Simulated from sample)
light_name = "Bedroom Light"
action_input = "ON"

# 2. Creating an instance of the class
my_light = SmartLight(light_name)

# 3. Performing the action
my_light.set_action(action_input)
print("--- Sample Output 1 (Expected) ---")

# 4. Displaying the final result
my_light.display_status()

# --- Sample 2 (Alternative Output) ---
light2 = SmartLight("Kitchen Light")
light2.set_action("OFF")
print("\n--- Sample Output 2 (Alternative) ---")
light2.display_status()
```

# **Problem 2: Employee ID Card System**

```python
class Employee:
    def __init__(self, name, emp_id, department):
        """Initialize employee with name, ID, and department."""
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        """Display the formatted ID card details."""
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

# --- Testing with Sample Input ---

# 1. Provide the input data
name_input = "Rahul"
id_input = "EMP102"
dept_input = "IT"

# 2. Create an instance (object) of the Employee class
emp1 = Employee(name_input, id_input, dept_input)
print("--- Sample Output 1 (Expected) ---")

# 3. Call the method to display the ID card
emp1.display_id_card()

# --- Sample 2 (Alternative Output) ---
emp2 = Employee("Priya", "EMP550", "HR")
print("\n--- Sample Output 2 (Alternative) ---")
emp2.display_id_card()
```

# **Problem 3: Mobile Contact Record**

```python
class Contact:
    def __init__(self, name, phone_number):
        """Initialize the contact with a name and phone number."""
        self.name = name
        self.phone_number = phone_number

    def display_contact(self):
        """Display the saved contact information in the required format."""
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone_number}")

# --- Testing with Sample Input ---

# Sample 1: Exact data from the problem statement
contact1 = Contact("Anita", "9876543210")

# Sample 2: A different contact example
contact2 = Contact("Rahul", "8899776655")

# Displaying results
print("--- Sample Output 1 (Expected) ---")
contact1.display_contact()

print("\n--- Sample Output 2 (Alternative) ---")
contact2.display_contact()
```

# **Problem 4: Product Price Tag Generator**

```python
class Product:
    def __init__(self, name, price):
        """Initialize the product with a name and price."""
        self.name = name
        self.price = price

    def display_tag(self):
        """Print the formatted price tag with the Rupee symbol."""
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")

# --- Testing with two samples ---

# Sample 1: Using the data from the problem statement
p1 = Product("Headphones", 2499)

# Sample 2: Using alternative data
p2 = Product("Smartphone", 15999)

# Displaying Results
print("--- Sample Output 1 (Expected) ---")
p1.display_tag()

print("\n--- Sample Output 2 (Alternative) ---")
p2.display_tag()
```

# **Problem 6: Delivery Address Manager**
