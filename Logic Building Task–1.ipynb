# **1. User Login Check**

```python
# Predefined correct credentials
correct_username = "admin"
correct_password = "1234"

# Taking input from the user
entered_username = input("Enter username: ")
entered_password = input("Enter password: ")

# Checking whether login is successful
if entered_username == correct_username and entered_password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

# **2. Pass / Fail Analyzer**

```python
# List of marks provided in the problem
marks = [45, 78, 90, 33, 60]

# Initializing counters for pass and fail students
pass_count = 0
fail_count = 0

# Iterating through the marks list
for m in marks:
    if m >= 50:
        pass_count += 1  # Increment pass counter if marks are 50 or more
    else:
        fail_count += 1  # Increment fail counter if marks are below 50

# Printing the final results clearly
print("--- Pass / Fail Analysis Results ---")
print(f"Total number of pass students: {pass_count}")
print(f"Total number of fail students: {fail_count}")
```

# **3. Simple Data Cleaner**

```python
# Original list of names with spaces and mixed casing
names = [" Alice ", "bob", " CHARLIE "]

# Creating a new list to store cleaned names
cleaned_names = []

# Iterating through each name in the list
for name in names:
    # .strip() removes leading and trailing spaces
    # .lower() converts the string to lowercase
    cleaned = name.strip().lower()
    cleaned_names.append(cleaned)

# Print the final cleaned list
print("Cleaned List:", cleaned_names)
```

```python
#Alternate Method

names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = [name.strip().lower() for name in names]
print("Cleaned Names: ",cleaned_names)
```

# **4. Message Length Analyzer**

```python
# List of messages provided in the problem
messages = ["Hi", "Welcome to the platform", "OK"]

print("--- Message Analysis ---")

# Iterating through each message in the list
for msg in messages:
    # 1. Calculate the length of the message
    msg_length = len(msg)

    # 2. Print the length of each message
    print(f"Message: '{msg}' | Length: {msg_length}")

    # 3. Flag messages longer than 10 characters
    if msg_length > 10:
        print("   -> FLAG: This message is longer than 10 characters.")
```

# **5. Error Message Detector**

```python
# List of log entries provided in the problem
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Initializing a counter for error entries
error_count = 0

# Iterating through each log entry
for entry in logs:
    # Check if the entry matches the string "ERROR"
    if entry == "ERROR":
        error_count += 1  # Increment counter if "ERROR" is found

# Print the total error count
print(f"Total error count: {error_count}")
```
